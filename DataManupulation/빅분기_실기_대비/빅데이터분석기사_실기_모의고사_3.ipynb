```python
#5.top year 년도를 기준으로 발매일(year released)과 top100에 진입한 
#일자 (top year)가 다른 곡의 숫자를 count 했을때 가장 많은 년도는?
df['year released'] =df['year released'].astype(int)
df['top year'] = df['top year'].astype(int)
print(df[(df['year released'] - df['top year']) != 0].groupby('top year').size().sort_values(ascending = False).index[0])
```
