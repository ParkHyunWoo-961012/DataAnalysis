```python
import requests
import pandas as pd
import datetime
from bs4 import BeautifulSoup

```

```python
url = "https://finance.naver.com/sise/sise_group.naver?type=upjong"
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')
soup = soup.find(class_="type_1").findAll('a')
```

```python
page_num = [] # for finding stock name
industry_name = []
for i in range(len(soup)):
    if soup[i].string != "기타" and soup[i].string !="창업투자": # not need '기타' category because that have futures(선물) 
        page_num.append(soup[i].get("href"))
        industry_name.append(soup[i].string)
```

```python
classfied_stock = {}
whole_stock_name = []
for i in range(len(page_num)):
    res = requests.get("https://finance.naver.com/" + page_num[i])
    soup = BeautifulSoup(res.text,'lxml')
    soup = soup.find('tbody').find_all('a')
    for j in range(len(soup)):
        stock_name = soup[j].getText()
        if stock_name != "":
            classfied_stock[stock_name] = industry_name[i]
            whole_stock_name.append(stock_name)            
```

```python
classfied_stock = pd.DataFrame(list(classfied_stock.items()),columns=['Stock_Name', 'Sectors'])
stock_code = pd.read_csv("stock_code.csv",encoding='CP949')
stock_code = stock_code[['단축코드','한글 종목약명']]
stock_code = stock_code[stock_code['한글 종목약명'].isin(whole_stock_name)]
stock_code = stock_code.rename(columns = {'한글 종목약명':'Stock_Name'})

classified_stock = classfied_stock.merge(stock_code,on='Stock_Name')
```

```python
stock_price = None
sector_first = None
headers = {'User-Agent' : 'My_header'}
sectors  = list(classified_stock['Sectors'].unique())
stock_name = classified_stock.loc[classified_stock['Sectors']==sectors[0]]['단축코드']

#for code in stock_name:
for code in stock_name:
    for page_num in range(1,20):
        url = "https://finance.naver.com/item/sise_day.naver?code="+code+"&page="+str(page_num)
        req = requests.get(url,headers =headers)
        stock_price = pd.concat([stock_price,pd.read_html(req.text,encoding = 'euc-kr')[0]])

    stock_price.dropna(inplace = True)

    fluctuation_list = []
    for i in range(0,len(stock_price)-1):
        present = stock_price['종가'].values[i]
        previous = stock_price['종가'].values[i+1]
        if(present>previous):
            fluctuation = round(stock_price['전일비'].values[i]*100/present,2)
            fluctuation_list.append(fluctuation)
        else:
            fluctuation = round(0-(stock_price['전일비'].values[i]*100/present),2)
            fluctuation_list.append(fluctuation)

    fluctuation_list.append(0)

    stock_price['변동률'] = fluctuation_list
    if code == stock_name[0]:
        sector_first = stock_price[['날짜','거래량','변동률']]
    else:
        sector_first['변동률'] = sector_first['변동률'].values+stock_price['변동률'].values
        sector_first['거래량'] = sector_first['거래량'].values+stock_price['거래량'].values
    
```

```python
sector_first["거래량"] = sector_first["거래량"]/len(stock_name)
sector_first["변동률"] = sector_first["변동률"]/len(stock_name)
```

```python
Sector_SC = None
Sector_Bio = None
Sector_Car = None
Sector_Gas = None
Sector_Game = None
sectors = ["반도체와반도체장비" , "제약" , "자동차" , "석유와가스" , "게임엔터테인먼트"]
target_sector = [Sector_SC,Sector_Bio,Sector_Car,Sector_Gas,Sector_Game]


for sector_order in range(5):
    stock_name = classified_stock.loc[classified_stock['Sectors']==sectors[sector_order]]['단축코드']
    stock_price = None

    for code in stock_name:
        for page_num in range(1,3):
            url = "https://finance.naver.com/item/sise_day.naver?code="+code+"&page="+str(page_num)
            req = requests.get(url,headers =headers)
            stock_price = pd.concat([stock_price,pd.read_html(req.text,encoding = 'euc-kr')[0]])

            stock_price.dropna(inplace = True)
            stock_price['target'] = ((stock_price['시가'].values + stock_price['종가'].values)/2)*stock_price['거래량'].values

            if code == stock_name.values[0] and page_num ==1:
                target_sector[sector_order] = stock_price
            else:
                target_sector[sector_order]['target'] = target_sector[sector_order]['target']+ stock_price['target']
            stock_price = None

    target_sector[sector_order]['target'] = target_sector[sector_order]['target']/len(stock_name)/1000000
```

```python
result = None
for i in range(5):
    target_sector[i].rename(columns = {'target' : sectors[i]}, inplace = True)
    if i==0:
        result = target_sector[i]
    else:
        result = pd.merge(result,target_sector[i][[sectors[i]]+["날짜"]],on="날짜")
        
        
result = result[["날짜"]+sectors]
```
