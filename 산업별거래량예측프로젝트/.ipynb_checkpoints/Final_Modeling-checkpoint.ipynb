```python
Target_Columns = data.columns[16:21]
Input_Columns = data.columns.difference(Target_Columns)
```

```python
train_data = data[:199]
test_data = data[199:]

Semi_Model = None
Bio_Model = None
Car_Model = None
Gas_Model = None
Game_Model = None

test_semi_Model = None
test_bio_Model = None
test_car_Model = None
test_gas_Model = None
test_game_Model = None

Model_Data = [Semi_Model,Bio_Model,Car_Model,Gas_Model,Game_Model]
test_model_data = [test_semi_Model,test_bio_Model,test_car_Model,test_gas_Model,test_game_Model]

for i in range(len(Target_Columns)):
    Model_Data[i] = train_data
    Model_Data[i]["Target"] = data[Target_Columns.values[i]]
    Model_Data[i] = Model_Data[i].drop(columns=Target_Columns.values[i])
    
    test_model_data[i] = test_data
    test_model_data[i]["Target"] = data[Target_Columns.values[i]]
    test_model_data[i] = test_model_data[i].drop(columns=Target_Columns.values[i])
```

```python
for i in range(len(Target_Columns)):
    Model_Data[i]["Group"] = 0
    thirdQ = Model_Data[i]["Target"].quantile(0.75)
    secondQ = Model_Data[i]["Target"].quantile(0.5)
    firstQ = Model_Data[i]["Target"].quantile(0.25)
    print(secondQ , firstQ, thirdQ)
    Model_Data[i].loc[Model_Data[i]['Target']>thirdQ,"Group"] = 0
    Model_Data[i].loc[(Model_Data[i]['Target']>secondQ) & (Model_Data[i]['Target']<thirdQ),"Group"] = 1
    Model_Data[i].loc[(Model_Data[i]['Target']>firstQ) & (Model_Data[i]['Target']<secondQ),"Group"] = 2
    Model_Data[i].loc[Model_Data[i]['Target']<firstQ,"Group"] = 3
```

```python
# 순서 ['반도체와반도체장비', '제약', '자동차섹터', '석유와가스', '게임엔터테인먼트']

Reg0 = None
Reg1 = None
Reg2 = None
Reg3 = None
Reg4 = None
coef_semi = None
coef_bio = None
coef_car = None
coef_gas = None
coef_game = None

vif_result = []
group_kfold_ensemble_score = []
ensemble_index_value = []
ensemble_real_value = []
ensemble_predict_value = []
ensemble = []
ensemble_score = []
ensemble_MSE = []
ensemble_MAE = []



lin_regression = [Reg0,Reg1,Reg2,Reg3,Reg4]
coef_df = [coef_semi,coef_bio,coef_car,coef_gas,coef_game]
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
pd.set_option('mode.chained_assignment',  None)
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import BaggingRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GroupKFold

lin_score = []
for i in range(len(Target_Columns.values)):
    Model_Data[i].index = Model_Data[i]['Date']
    Model_Data[i].drop(columns = "Date",inplace=True)
    X = Model_Data[i][Model_Data[i].columns[:-2]]
    y = Model_Data[i]["Target"]
    group = Model_Data[i]["Group"]
    
    test_model_data[i].index = test_model_data[i]['Date']
    test_model_data[i].drop(columns = "Date",inplace=True)
    
    test_X = test_model_data[i][test_model_data[i].columns[:-1]]
    test_y = test_model_data[i][test_model_data[i].columns[-1:]]
    
    X_train, X_val, Y_train, Y_val = train_test_split(X, y, test_size=0.25, shuffle=False)

    # 앙상블 모델 생성
    Reg = LinearRegression()
    bagging = BaggingRegressor(base_estimator = Reg)
    bagging.fit(X,y)
    ensemble.append(bagging)
    ensemble_score.append(bagging.score(X,y))
#     ensemble_index_value.append(test_y.index)
#     ensemble_real_value.append(test_model_data[i]["Target"])
#     ensemble_predict_value.append(bagging.predict(test_X))
    ensemble_index_value.append(Y_val.index)
    ensemble_real_value.append(Y_val)
    ensemble_predict_value.append(bagging.predict(X_val))
    
    ensemble_MSE.append(mean_squared_error(Y_val,bagging.predict(X_val)))
    ensemble_MAE.append(mean_absolute_error(Y_val,bagging.predict(X_val)))
    
    for j in range(len(bagging.estimators_)):
        if j == 0:
            coef = bagging.estimators_[j].coef_
        else:
            coef = coef + bagging.estimators_[j].coef_
        
    coef = coef/5
    coef_df[i] = pd.DataFrame(index = Model_Data[i].columns[:-2])
    coef_df[i]["Coefficient"] = coef
    
    
    gkf = GroupKFold(n_splits=group.nunique())

    score = []
    for train_idx, val_idx in gkf.split(X,y,np.array(group.values)):
        train_X, val_X = X.iloc[train_idx],X.iloc[val_idx]
        train_y, val_y = y.iloc[train_idx],y.iloc[val_idx]
        bagging.fit(train_X,train_y)
        score.append(bagging.score(val_X,val_y))
     
    group_kfold_ensemble_score.append(score)
```

```python
def scaler(data):
    std = data.std()
    mean = data.mean()
    new_value = [] 
    for i in range(len(data)):
        std_data = (data[i]- mean)/std
        new_value.append(std_data)
        
    data.replace(data.values,new_value,inplace=True)

for i in data.columns.values:
    if i != "date":
        scaler(data[i])
```

```python
std_Semi_Model = None
std_Bio_Model = None
std_Car_Model = None
std_Gas_Model = None
std_Game_Model = None

std_Model_Data = [std_Semi_Model,std_Bio_Model,std_Car_Model,std_Gas_Model,std_Game_Model]

for i in range(len(Target_Columns)):
    std_Model_Data[i] = data
    std_Model_Data[i]["Target"] = data[Target_Columns.values[i]]
    std_Model_Data[i] = std_Model_Data[i].drop(columns=Target_Columns.values[i])
```

```python
# 순서 ['반도체와반도체장비', '제약', '자동차섹터', '석유와가스', '게임엔터테인먼트']

std_Reg0 = None
std_Reg1 = None
std_Reg2 = None
std_Reg3 = None
std_Reg4 = None 
std_ols_summary = []
std_vif_result = []
std_lin_score = []
std_kfold_lin_score = []
std_MSE_value = []
std_MAE_value = []
std_ensemble = []
std_ensemble_score = []
std_lin_regression = [std_Reg0,std_Reg1,std_Reg2,std_Reg3,std_Reg4]
```

```python
for i in range(len(Target_Columns.values)):
    X = std_Model_Data[i][std_Model_Data[i].columns[:-1]]
    X.drop(columns = 'date',inplace=True)
    y = std_Model_Data[i][std_Model_Data[i].columns[-1:]]
 
    kfold = KFold(n_splits=5,shuffle=True)
    std_linear_score = []
    
    
    Reg = LinearRegression()
    for train_idx, val_idx in kfold.split(X):
        train_X, val_X = X.iloc[train_idx],X.iloc[val_idx]
        train_y, val_y = y.iloc[train_idx],y.iloc[val_idx]

        Reg.fit(train_X,train_y)
        std_linear_score.append(Reg.score(val_X,val_y))
    
    std_kfold_lin_score.append(std_linear_score)
    Reg = LinearRegression()
    Reg.fit(X,y)
    std_lin_regression[i] = Reg
    std_lin_score.append(Reg.score(X,y))
    
    std_MSE_value.append(mean_squared_error(y, Reg.predict(X)))
    std_MAE_value.append(mean_absolute_error(y, Reg.predict(X)))
    
    X_train, X_val, Y_train, Y_val = train_test_split(X, y, test_size=0.25, random_state=321)

    Reg = LinearRegression()
    bagging = BaggingRegressor(base_estimator = Reg,oob_score = True)
    bagging.fit(X_val,Y_val)
    std_ensemble.append(bagging)
    std_ensemble_score.append(bagging.score(X_val,Y_val))
       
    X = sm.add_constant(X)
    model = sm.OLS(y,X)
    result = model.fit()
    std_ols_summary.append(result.summary())

    vif = pd.DataFrame()
    vif["VIF value"] = [variance_inflation_factor(X.values,i) for i in range(0,len(X.columns))]
    vif["explanatory variables"] = X.columns
    std_vif_result.append(vif)

```

```python
# all data is standardizaitoned
for i in range(5):
    print("Linear Score : "+ Target_Columns[i] + " sector : " + str(std_lin_score[i]))
    print("Kfold Average Score : "+ Target_Columns[i] + " sector : " + str(np.mean(std_kfold_lin_score[i])))
    print("Ensemble Score : "+ Target_Columns[i] + " sector : " + str(std_ensemble_score[i])+"\n\n")
    
```
